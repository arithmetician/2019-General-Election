###  Relationship votes between Geopolitical zone and Accredited voters

```python
# Group by Geopolitical zone and Accredited Voters

vote_count =data.groupby(['Geopolitical zone','Accredited Voters'])['Valid votes'].sum().reset_index()
```

```python
vote_count
```

```python
# Get the heighest vote in each zone 

heighest_vote = vote_count.loc[vote_count.groupby('Geopolitical zone')['Accredited Voters'].idxmax()]
```

```python
heighest_vote
```

```python
# Visualization and its libraries

import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.bar(heighest_vote['Geopolitical zone'] ,heighest_vote['Accredited Voters'])
plt.title('Heighest Vote by Geopolitical Zone')
plt.xlabel('Geopolitical zone')
plt.ylabel('Accredited Voters')

plt.show()
```

```python
# Group  data by PDP and APC

vote_counts =data.groupby(['PDP Votes','APC Votes'])['Others'].sum().reset_index()
```

```python
vote_counts
```

```python
# Get the heighest between the parties 

heighest_votes = vote_counts.loc[vote_counts.groupby('PDP Votes')['APC Votes'].idxmax()]
```

```python
heighest_votes
```

```python
plt.figure(figsize=(10,8))
plt.scatter(heighest_votes['PDP Votes'] ,heighest_votes['APC Votes'] )
plt.title('Heighest Vote by PDP and APC')
plt.xlabel('PDP votes')
plt.ylabel('APC Votes')

plt.show()
```
